```python
def find_conj_words_list():
    matcher = Matcher(nlp.vocab)
    conj_list =[]
    pattern = [{'POS':'NOUN'}, {'POS': 'CCONJ'},{'POS':'NOUN'}]
    matcher.add("conj",[pattern])
    matches = matcher(article)
    for match_id, start, end in matches:
        span = article[start:end]
        conj_list.append([get_full_word(span[0]),span[1],get_full_word(span[-1])])
    return conj_list
```

```python
def find_rel_location():
    matcher = Matcher(nlp.vocab)
    rel_list =[]
    pattern = [{'ENT_TYPE':'ORG'}, {'POS': {'IN':['ADP','PUNCT']}},{'POS': 'DET', 'OP': '?'},{'ENT_TYPE':'GPE'}]
    matcher.add("located",[pattern])
    matches = matcher(article)
    for match_id, start, end in matches:
        span = article[start:end]
        
        rel_list.append([get_full_word(span[0]),'located in',get_full_word(span[-1])])
    return rel_list
find_rel_location()
```
