```python
def get_relation(sent):

    doc = nlp(sent)

    # Matcher class object 
    matcher = Matcher(nlp.vocab)

    #define the pattern 
    pattern = [#{'DEP':'ROOT'}, 
            {'DEP':'prep','OP':"?"},
            {'DEP':'agent','OP':"?"},  
            {'POS':'VERB'}
#             {'POS':'ADJ','OP':"?"}
    ] 

    matcher.add("matching_1", [pattern]) 

    matches = matcher(doc)
    k = len(matches) - 1
    print(k)
    span = doc[matches[k][1]:matches[k][2]] 
#     print(matches)
    rel = [doc[matches[k][1]:matches[k][2]] for k in range(len(matches))]
    return rel
```

```python
def extract_rel_match(doc, matcher):
    for sent in doc.sents:
#         print(sent)
        for match_id, start, end in matcher(sent):
#             print("test")
            span = sent[start:end]  # matched span
            pred = nlp.vocab.strings[match_id] # rule name
            subj, obj = span[0], span[-1]
#             print(span)
            if pred.startswith('rev-'): # reversed relation
                subj, obj = obj, subj
                pred = pred[4:]
            print(subj)
            print(obj)
#             yield ((subj._.ref_n, subj._.ref_t), pred, 
#                    (obj._.ref_n, obj._.ref_t))
            yield ((subj, subj), pred, 
                   (obj, obj))
```
